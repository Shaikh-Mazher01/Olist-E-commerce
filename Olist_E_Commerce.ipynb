*   order_id: The unique primary identifier for each individual transaction.
*   customer_id: The unique token linking the order to a specific customer's session (Foreign Key matching the customers table).
*   order_status: Current status of order wherther it is delivered or being proceesed or many more.

```python
non_delivered_statuses = ['canceled', 'processing', 'shipped', 'invoiced', 'approved', 'created','unavailable']
non_delivered_orders = orders[orders['order_status'].isin(non_delivered_statuses)]
non_delivered_orders
```

```python
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'], format='%Y-%m-%d %H:%M:%S')
orders['order_approved_at'] = pd.to_datetime(orders['order_approved_at'], format='%Y-%m-%d %H:%M:%S')
orders['order_delivered_carrier_date'] = pd.to_datetime(orders['order_delivered_carrier_date'], format='%Y-%m-%d %H:%M:%S')
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'], format='%Y-%m-%d %H:%M:%S')
orders['order_estimated_delivery_date'] = pd.to_datetime(orders['order_estimated_delivery_date'], format='%Y-%m-%d %H:%M:%S')
```

```python
delivered_mask = orders['order_status'] == 'delivered'

orders['delivery_delay'] = np.where(
    delivered_mask & (orders['order_delivered_customer_date'] > orders['order_estimated_delivery_date']),
    1, 0
)
orders.loc[~delivered_mask, 'delivery_delay'] = -1
```

```python
orders['delivery_duration'] = orders['order_delivered_customer_date'] - orders['order_purchase_timestamp']
```

```python
orders['month'] = orders['order_purchase_timestamp'].dt.to_period('M').astype(str)
```

```python
orders['weekday'] = orders['order_purchase_timestamp'].dt.day_name()
```

```python
orders['hour'] = orders['order_purchase_timestamp'].dt.hour
```

```python
orders['days'] = orders['delivery_duration'].dt.total_seconds() / 86400
```

```python
orders.duplicated().sum()
```

```python
orders.duplicated(subset=['order_id']).sum()
```

```python
orders.isnull().sum()
```

I have cleaned the data by converting the relevant columns to the timestamp data type. I then created a delivery_delay and delivery_duration column to check for any delayed orders and time duration of delivery respectively. As delivery_duration is in Timedelta format so break down into days,month,weekdays and hour, also verified that there are no duplicate values in order_id. The dataset is now fully cleaned.

```python
manual_translator = {
    'pc_gamer': 'pc_gamer',
    'portateis_cozinha_e_preparadores_de_alimentos': 'portable_kitchen_food_preparers'
}
```

```python
products = products.drop(columns=['product_category_name'])
products = products.rename(columns={'product_category_name_english': 'product_category'})
products[['product_id','product_category']]
```

```python
products.isnull().sum()
```

Here 610 product name,product description and product category is missing let to handle this fill with median length of product and description and category as other and drop those teo rows having missing dimension and weight.

```python
products['product_category'] = products['product_category'].fillna('other')
```

```python
temp_list = ['product_name_lenght', 'product_description_lenght', 'product_photos_qty']

for col in temp_list:
    col_median = products[col].median()
    products[col] = products[col].fillna(col_median).infer_objects(copy=False)
```

```python
dimension = ['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']
products = products.dropna(subset=dimension)
```

```python
payments_agg = order_payments.groupby('order_id').agg(
    total_payment_value=('payment_value', 'sum'),
    max_installments=('payment_installments', 'max'),
    payment_type=('payment_type', lambda x: x.mode()[0] if not x.mode().empty else 'unknown')
).reset_index()

print("payments_agg shape:", payments_agg.shape)
```

```python
df_profitability = order_items.merge(orders, on='order_id', how='inner')
df_profitability = df_profitability.merge(customers, on='customer_id', how='left')

state_metrics = df_profitability.groupby('customer_state').agg(
    total_revenue=('price', 'sum'),
    total_freight=('freight_value', 'sum'),
    order_count=('order_id', 'nunique'),
    total_items=('order_item_id', 'count')
).reset_index()

# Because of population bias created relative matrix
state_metrics['Average_Order_Value'] = state_metrics['total_revenue'] / state_metrics['order_count']
state_metrics['Freight_Overhead_Percent'] = (state_metrics['total_freight'] / state_metrics['total_revenue']) * 100

state_metrics = state_metrics[state_metrics['order_count'] > 50]

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

state_metrics_aov = state_metrics.sort_values(ascending=False, by='Average_Order_Value')
sns.barplot(
    data=state_metrics_aov,
    x='Average_Order_Value',
    y='customer_state',
    palette='coolwarm',
    hue='customer_state',
    ax=axes[0]
)
axes[0].set_title('Average Order Value (AOV) by State\n(Where do customers spend the most per order?)', fontsize=12)
axes[0].set_xlabel('Average Spend per Order (BRL)')
axes[0].set_ylabel('Customer State')

state_metrics_freight = state_metrics.sort_values(ascending=True, by='Freight_Overhead_Percent')
sns.barplot(
    data=state_metrics_freight,
    x='Freight_Overhead_Percent',
    y='customer_state',
    palette='viridis',
    hue='customer_state',
    ax=axes[1]
)
axes[1].set_title('Freight Overhead Ratio by State\n(Lower % = Better Profit Margins)', fontsize=12)
axes[1].set_xlabel('Freight Cost as % of Product Price')
axes[1].set_ylabel('')

plt.tight_layout()
plt.show()
```

Average Order Value (AOV) by State (Left Plot): Normalizes regional consumer habits, revealing that remote or less-populated states frequently yield a significantly higher average dollar spend per individual transaction compared to industrialized urban centers.

Freight Overhead Ratio by State (Right Plot): Highlights geographical margin pressures by proving that remote regions incur highly defensive shipping costs (exceeding 30% of product price), whereas infrastructure-heavy distribution hubs drop overhead below 15%.
